```python
# writting functions for previous code
import matplotlib.pyplot as plt
import tensorflow as tf

def gabor(n_values=32, sigma=1.0, mean=0.0):
    k = tf.linspace(-3.0, 3.0, n_values)
    z = (tf.exp(tf.negative(tf.pow(k - mean, 2.0) / (2.0 * tf.pow(sigma, 2.0)))) * (1.0 / (sigma * tf.sqrt(2.0 * 3.1415))))
    gauss_kernel = tf.matmul(tf.reshape(z, [n_values, 1]), tf.reshape(z, [1, n_values]))
    x = tf.reshape(tf.sin(k), [n_values, 1])
    y = tf.reshape(tf.ones_like(x), [1, n_values])
    gabor_kernel = tf.matmul(tf.matmul(x, y), gauss_kernel)
    return gabor_kernel
```

```python
# writting convolve function

import logging

def convolve(img, W):
    # the W matrix is only 2D
    # but conv2d need 4D tensor
    # height * width * n_input * n_output
    if len(W.get_shape()) == 2:
        dims = W.get_shape().as_list() + [1,1]
        W = tf.reshape(W, dims)
        
    if len(img.get_shape()) == 2:
        dims = [1] + img.get_shape().as_list() + [1]
        tf.reshape(img, dims)
    elif len(img.get_shape()) == 3: 
        dims = [1] + img.get_shape().as_list()
        img = tf.reshape(img, dims)
        # if the image is 3 channels, then our convolution
        # kernel needs to be repeated for each input channel
        W = tf.concat([W,W,W], axis=2)
        
    # Stride is how many values to skip for the dimensions of
    # num, height, width, channels
    img = tf.cast(img, tf.float32)
    convolved = tf.nn.conv2d(img, W, strides=[1,1,1,1], padding='SAME')
    return convolved
```

```python
from skimage import data
img = tf.convert_to_tensor(data.astronaut())
plt.imshow(img)
plt.show()
print(img.shape)
```

```python
def gabor_convolve_image(x):
    return tf.squeeze(convolve(x, gabor()))
```

```python
result = gabor_convolve_image(img)
plt.imshow(result)
plt.show()
```
