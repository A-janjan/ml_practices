```python
import numpy as np
import matplotlib.pyplot as plt

def guassian(x,  mu, sig):
    return np.exp((- ( np.power( (x-mu) , 2. ) ) )/(2*np.power(sig, 2.)))
```

```python
x_vals = np.linspace(-3,3,100)

for mu, sig in [(-1,1), (0,2), (2,3)]:
    y_vals = [guassian(val, mu, sig) for val in x_vals]
    plt.plot(x_vals, y_vals)
plt.show()
```

```python
import tensorflow as tf

class GaussianModule(tf.Module):
    def __init__(self):
        self.mu = tf.Variable(1.0, name='mu')
        self.sig = tf.Variable(1.0, name='sig')  # Initialize sigma with a positive value
    
    def __call__(self, x):
        return tf.exp(-tf.pow(x - self.mu, 2) / (2 * tf.pow(self.sig, 2.)))

def compute_loss(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true - y_pred))  # Example loss function (Mean Squared Error)

# Create an instance of the GaussianModule
model = GaussianModule()

# Define an optimizer
optimizer = tf.optimizers.Adam(learning_rate=0.01)

@tf.function
def train_step(x, y):
    with tf.GradientTape() as tape:
        y_pred = model(x)
        loss = compute_loss(y, y_pred)
    gradients = tape.gradient(loss, [model.mu, model.sig])  # Include both mu and sig in the gradients
    optimizer.apply_gradients(zip(gradients, [model.mu, model.sig]))
    return loss

# Example usage
x = tf.constant([1.0, 2.0, 3.0], dtype=tf.float32)
y = tf.constant([1.0, 2.0, 3.0], dtype=tf.float32)

for step in range(100):
    loss = train_step(x, y)
    print(f"Step {step}, Loss: {loss.numpy()}, mu: {model.mu.numpy()}, sig: {model.sig.numpy()}")
```

```python
mu_learned = model.mu.numpy()
sig_learned = model.sig.numpy()

y_learned = tf.exp(-tf.pow(x-mu_learned, 2.)/(2*tf.pow(sig_learned, 2.)))

plt.plot(x, y_learned)
plt.show()
```
